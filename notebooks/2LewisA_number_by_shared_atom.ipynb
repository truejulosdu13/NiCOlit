```python
from rdkit import Chem
from rdkit.Chem import Draw
import numpy as np
import pandas as pd
import sys
sys.path.append('../')
import dft_descriptors.prepocessing as pp
from dft_descriptors import numbering_CO as nb
```

```python
df = pd.read_csv('../data_csv/Data_test11222021.csv', sep = ',')
df = pp.preprocess(df)
df = df[df["Lewis Acid"].notna()]
```

```python
L = list(df["Lewis Acid"].unique())
L.remove('nan')
L.remove('NoLewisAcid')
len(L)
```

```python
#### determine if ligand is NHC, Phosphine, Diphosphine or other ####
def number_AL(list_lig):
    num_als = []
    for smi in list_lig:
        m = Chem.MolFromSmiles(smi)
        if m.HasSubstructMatch(Chem.MolFromSmarts('[Mg]')):
            smi_num = Chem.MolToSmiles(number_Mg(smi))
        elif m.HasSubstructMatch(Chem.MolFromSmarts('[Zn]')):
            smi_num = Chem.MolToSmiles(number_Zn(smi))  
        elif m.HasSubstructMatch(Chem.MolFromSmarts('[Al]')):
            smi_num = Chem.MolToSmiles(number_Al(smi)) 
        elif m.HasSubstructMatch(Chem.MolFromSmarts('B')):
            smi_num = Chem.MolToSmiles(number_B(smi)) 
        elif m.HasSubstructMatch(Chem.MolFromSmarts('[Li]')):
            smi_num = Chem.MolToSmiles(number_Li(smi)) 
        elif m.HasSubstructMatch(Chem.MolFromSmarts('[Na]')):
            smi_num = Chem.MolToSmiles(number_Na(smi)) 
        elif m.HasSubstructMatch(Chem.MolFromSmarts('[K]')):
            smi_num = Chem.MolToSmiles(number_K(smi)) 
        else:
            smi_num = Chem.MolToSmiles(number_M(smi)) 
            print(smi_num)
        num_als.append(smi_num)
    return num_als

#### RMgX ####
def number_Mg(smi):
    m = Chem.MolFromSmiles(smi)
    for at in m.GetAtoms():
        if at.GetSymbol() == 'Mg':
            nb.reset_atom_map(m, 0)
            at.SetAtomMapNum(0)
    return m

#### Zincate ####
def number_Zn(smi):
    m = Chem.MolFromSmiles(smi)
    for at in m.GetAtoms():
        if at.GetSymbol() == 'Zn':
            nb.reset_atom_map(m, 0)
            at.SetAtomMapNum(0)
    return m

#### Lithium ####
def number_Li(smi):
    m = Chem.MolFromSmiles(smi)
    for at in m.GetAtoms():
        if at.GetSymbol() == 'Li':
            nb.reset_atom_map(m, 0)
            at.SetAtomMapNum(0)
    return m

#### Aluminum ####
def number_Al(smi):
    m = Chem.MolFromSmiles(smi)
    for at in m.GetAtoms():
        if at.GetSymbol() == 'Al':
            nb.reset_atom_map(m, 0)
            at.SetAtomMapNum(0)
    return m

#### Boron ####
def number_B(smi):
    m = Chem.MolFromSmiles(smi)
    for at in m.GetAtoms():
        if at.GetSymbol() == 'B':
            nb.reset_atom_map(m, 0)
            at.SetAtomMapNum(0)
    return m

#### Sodium ####
def number_Na(smi):
    m = Chem.MolFromSmiles(smi)
    for at in m.GetAtoms():
        if at.GetSymbol() == 'Na':
            nb.reset_atom_map(m, 0)
            at.SetAtomMapNum(0)
    return m

#### Potassium ####
def number_K(smi):
    m = Chem.MolFromSmiles(smi)
    for at in m.GetAtoms():
        if at.GetSymbol() == 'K':
            nb.reset_atom_map(m, 0)
            at.SetAtomMapNum(0)
    return m

#### Other metals ####
def number_M(smi):
    m = Chem.MolFromSmiles(smi)
    for at in m.GetAtoms():
        if at.GetSymbol() == 'Sc':
            nb.reset_atom_map(m, 0)
            at.SetAtomMapNum(0)
        elif at.GetSymbol() == 'Ti':
            nb.reset_atom_map(m, 0)
            at.SetAtomMapNum(0)
        elif at.GetSymbol() == 'Rb':
            nb.reset_atom_map(m, 0)
            at.SetAtomMapNum(0)
        elif at.GetSymbol() == 'Cu':
            nb.reset_atom_map(m, 0)
            at.SetAtomMapNum(0)
        elif at.GetSymbol() == 'Ag':
            nb.reset_atom_map(m, 0)
            at.SetAtomMapNum(0)
        elif at.GetSymbol() == 'Fe':
            nb.reset_atom_map(m, 0)
            at.SetAtomMapNum(0)
        elif at.GetSymbol() == 'Sr':
            nb.reset_atom_map(m, 0)
            at.SetAtomMapNum(0)
        elif at.GetSymbol() == 'Y':
            nb.reset_atom_map(m, 0)
            at.SetAtomMapNum(0)
        elif at.GetSymbol() == 'Zr':
            nb.reset_atom_map(m, 0)
            at.SetAtomMapNum(0)   
        elif at.GetSymbol() == 'Si':
            nb.reset_atom_map(m, 0)
            at.SetAtomMapNum(0) 
        elif at.GetSymbol() == 'Cs':
            nb.reset_atom_map(m, 0)
            at.SetAtomMapNum(0) 
        elif at.GetSymbol() == 'N' and at.GetFormalCharge() == 1:
            nb.reset_atom_map(m, 0)
            at.SetAtomMapNum(0)             
    return m
```

```python
num_als = number_AL(L)
```

```python
num_df  = pd.DataFrame({'react' : L,
                        'C0C7_num' : num_als},
                        columns = ['react', 'C0C7_num'])
```
